```python
#Apply the KNN Algorithm to the dataset without applying scaling operation and
#try to compare the results before scaling and after scaling.
```

```python
#with scaling
X=df.drop(['id','diagnosis'],axis=1)
y=df.diagnosis
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train=pd.DataFrame(scaler.fit_transform(X_train))
```

```python
scaled_X_train
```

```python
scaled_X_test=pd.DataFrame(scaler.transform(X_test))
scaled_X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6,metric='euclidean')
knn.fit(scaled_X_train,y_train)
```

```python
y_train_pred=knn.predict(scaled_X_train)
y_train_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred))
```

Now we can see both the classifications done with and without scaling, we see that the diagnosis with B and M in no scaling data has less precision,recall and f1-score when compared to that of in scaling data. The main aspect we consider is accuracy and clearly scaling data has more accuracy than non-scaled data. Even with the macro avg and weighted average the scaling data is way ahead of non scaled data, so there is no aspect where we can say that non-scaled data is better than scaled data, so we can conclude that scaling is necessary for classification with KNN.
