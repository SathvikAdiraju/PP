```python
# OneHotEncoder -> similar to pd.get_dummies
# Import one hot encoder
# create object 
# use column transform ,inside this we keep onehotencoder
# 2 mandatory arguments transforms and remainder , accepts a list of tuple values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
column_transform=ColumnTransformer(transformers=[('encode',OneHotEncoder(),[0])],remainder='passthrough')
#using remainder as passthrogh it continues till end
```

```python
#objectname.fit_transform(data)
```

```python
df=column_transform.fit_transform(df)
```

```python
df
```

```python
X=pd.DataFrame(df)
```

```python
X
```

```python
column_transform1=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='drop')
#using remainder as drop we drop after encoding
```

```python
df=column_transform1.fit_transform(df)
```
