Read an image(which contains different objects) using matplotlib and check the shape of the image?<br>
Display the image?<br>
Identify the different objects in the image?<br> 
Use slicing to get the objects and display them?<br>


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("download.jpg")
print(img.shape)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[:,:,0])
```

```python
plt.imshow(img[:,:,0],cmap='gray')
```

```python
plt.imshow(img[:,0:150])
```

```python
plt.imshow(img[:25,:100])
```

```python
plt.imshow(img[50:150,200:250])
```
